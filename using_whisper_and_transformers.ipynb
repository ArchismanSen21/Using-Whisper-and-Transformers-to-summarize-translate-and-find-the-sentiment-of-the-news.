*   Punkt Sentence Tokenizer

This tokenizer divides a text into a list of sentences by using an unsupervised algorithm to build a model for abbreviation words, collocations, and words that start sentences.

*   With the help of **nltk.tokenize.LineTokenizer**() method, we are able to extract the tokens from string of sentences in the form of single line by using tokenize.LineTokenizer() method.


*  **nltk.tokenize.sent_tokenize** return a sentence-tokenized copy of text, using NLTK’s recommended sentence tokenizer 





```python
import torch
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize
from nltk.tokenize import LineTokenizer
import math
from transformers import pipeline
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM


tokenizer = AutoTokenizer.from_pretrained('t5-base')
model = AutoModelForSeq2SeqLM.from_pretrained('t5-base')
device = torch.device("cuda")
model.to(device)
```

# Sentence Summarization using T5


using nltk.tokenizer we divide the sentence we got from audio into 8 chunks and each chunk is summerized using t5-base the result is then concatenated.

```python
lt = LineTokenizer()
chunk = 8 
```

```python
paragraphs = lt.tokenize(txt)   
summarised_para = []
```

```python
for paragraph in paragraphs:
    sentences = sent_tokenize(paragraph)
    chunkes = math.ceil(len(sentences) / chunk)     
    summary = []
    for i in range(chunkes):
        sent_batch = sentences[i*chunk:(i+1)*chunk]
        model_inputs = tokenizer(sent_batch, return_tensors="pt", padding=True, truncation=True, max_length=500).to(device)
        with torch.no_grad():
            summarised_batch = model.generate(**model_inputs)
        summary += summarised_batch
    summary = [tokenizer.decode(s, skip_special_tokens=True) for s in summary]
    summarised_para += [" ".join(summary)]

summarised_txt = "\n".join(summarised_para)
```

```python
print(summarised_txt)
print(len(summarised_txt))
```

# sentiment Analysis

**Model Description**

This model is a fine-tune checkpoint of DistilBERT-base-uncased, fine-tuned on SST-2. This model reaches an accuracy of 91.3 on the dev set (for comparison, Bert bert-base-uncased version reaches an accuracy of 92.7).


[https://huggingface.co/distilbert-base-uncased-finetuned-sst-2-english](https://)

```python
# Sentiment analysis pipeline
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
analyzer = pipeline("sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", tokenizer=tokenizer)
```

```python
sentiment = analyzer(summarised_txt)
print(sentiment)
```

# Bangla Translation

As i'm a bengali i want to read the summarized news in my mother tongue. i used a english to bengali transformer to translate the text from english to bengali.


[https://huggingface.co/csebuetnlp/banglat5_nmt_en_bn](https://)


```python
# Bangla translation pipeline
from transformers import AutoModelForSeq2SeqLM
from normalizer import normalize

model = AutoModelForSeq2SeqLM.from_pretrained("csebuetnlp/banglat5_nmt_en_bn")
tokenizer = AutoTokenizer.from_pretrained("csebuetnlp/banglat5_nmt_en_bn",use_fast=False)
```

```python
input_sentence = summarised_txt
input_ids = tokenizer(normalize(input_sentence), return_tensors="pt").input_ids
generated_tokens = model.generate(input_ids)
decoded_tokens = tokenizer.batch_decode(generated_tokens)[0]

print(decoded_tokens)
```
